```python
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
```

```python
def read_planning_log(suffix):
    file_path = "./planning/planning{}.log".format(suffix)
    with open(file_path) as planning_log_file:
        lines = planning_log_file.readlines()
        return np.array([float(line.strip().split()[-1]) for line in lines if "Planning response time" in line])
```

```python
def plot_planning_log(response_times):
    _, axes = plt.subplots(figsize=(20,10))
    axes.plot(range(len(response_times)), response_times)
    plt.show()
```

```python
def plot_planning_logs(rt_1, rt_2):
    _, axes1 = plt.subplots(figsize=(20,10))
    axes1.plot(range(len(rt_1)), rt_1)
    axes1.plot(range(len(rt_2)), rt_2)
    plt.show()
```

```python
planning_response_times = read_planning_log("")
plot_planning_log(planning_response_times)
planning_response_cho_6_times = read_planning_log("_cho_6")
plot_planning_log(planning_response_cho_6_times)
plot_planning_logs(planning_response_times, planning_response_cho_6_times)
print(len(planning_response_times), len(planning_response_cho_6_times))
```
